```python
import os
import pandas as pd
from math import isnan,ceil

cor_sheet_path = r'C:\Users\oussamaJB\Desktop\Full data\Sheet de correspondance.csv'
annot_sheets_path = r'C:\Users\oussamaJB\Desktop\Full data\annots'

# Opening the sheet of corresponding names
seq_df = pd.read_csv(cor_sheet_path)
seq_df['original name'] = seq_df['original name'].apply(lambda name:name[:name.index('.')-4]) 
seq_df.set_index('original name',inplace=True)
#print(seq_df)

# Opening the annotations sheets and extracting the required annotations
files = os.listdir(annot_sheets_path)
#print(files)
annots = {} 
count = 0
for file in files:
    if file[0]=='~':
        continue
    annot = pd.read_excel(annot_sheets_path+'/'+file,header=None,index_col=0)
    #print(annot.index)
    pb = list(annot.loc['2PB'])
    pn1 = list(annot.loc['1PN'])
    pn2 = list(annot.loc['2PN'])
    pn3 = list(annot.loc['3PN'])
    pnf = list(annot.loc['PN Fading'])
    ras = list(annot.loc['ras'])
    #print(ras)
    #print(pb)
    #print(file)
    sheet_id = file[file.index('#')+2:file.find('summary')-1]
    #print(sheet_id)
    for i in range(1,len(pb)+1):
        #print(pb[i])
        if isnan(pb[i-1]) and isnan(pn1[i-1]) and isnan(pn2[i-1]) and isnan(pn3[i-1]) and isnan(ras[i-1]) :
            print('skipped {} P{}'.format(sheet_id,i))
            #print(sheet_id)
            #print(pb)
            #print(pn)
            #print(pnf)
            continue
        else:
            sid = ''
            if i<10:
                sid = sheet_id+'-P0'+str(i)
            else:
                sid = sheet_id+'-P'+str(i)
            if sid not in seq_df.index:
                count +=1
                #print(sid) Herecount denotes the count of annotated data which we are not using due to poor quality
                continue
            if isnan(pb[i-1]):
                pb_frame = 0
            else:
                pb_frame = ceil(pb[i-1]*12)
            if isnan(pn2[i-1]):
                if isnan(pn1[i-1]):
                    if isnan(pn3[i-1]):
                        pn_frame = 0
                    else: 
                        pn_frame = ceil(pn3[i-1]*12)
                else:
                    pn_frame = ceil(pn1[i-1]*12)
            else:
                pn_frame = ceil(pn2[i-1]*12)
            if isnan(pnf[i-1]):
                pnf_frame = 0
            else:
                pnf_frame = ceil(pnf[i-1]*12)
            if not (isnan(ras[i-1])):
                pb_frame = 0
                pn_frame = 0
                pnf_frame = 0
            annots[ seq_df.loc[sid]['new name'] ] = {
                '2PB':pb_frame, '2PN':pn_frame, '2PNf': pnf_frame
            }
print(len(annots))
print(count)
```
